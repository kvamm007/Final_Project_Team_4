```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df = pd.read_csv('shark_attacks_cleaned.csv')
df.head()
```

```python
df = df.dropna()
df
```

```python
df['Fatal'] = df['Fatal'].str.replace('UNKNOWN','0')
df['Fatal'] = df['Fatal'].str.replace('N','0')
df['Fatal'] = df['Fatal'].str.replace('Y','1')
df
```

```python
X = pd.get_dummies(df, columns=['Date','Type','Country','Area','Location','Activity','Gender','Age'])
X = X.drop(columns=['Year','Fatal','Time','href formula','ISO_code'], axis=1)
y = df['Fatal']
```

```python
# Split training/test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```
