```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf

#  Import and read the csv.
attacks_df = pd.read_csv("shark_attacks_cleaned.csv")
attacks_df.head()
```

```python
attacks_df.drop(['index', 'Date', 'Time', 'href formula', 'ISO_code'], axis=1, inplace=True)
attacks_df
```

```python
attacks_df = attacks_df.dropna()
attacks_df
```

```python
attacks_df['Fatal'] = attacks_df['Fatal'].str.replace('UNKNOWN','0')
attacks_df['Fatal'] = attacks_df['Fatal'].str.replace('N','0')
attacks_df['Fatal'] = attacks_df['Fatal'].str.replace('Y','1')
attacks_df
```

```python
X = pd.get_dummies(attacks_df, columns=['Year','Type','Country','Area','Location','Activity','Gender','Age'])
y = attacks_df['Fatal']
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=2000,
                                random_state=1)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, predictions))
```
