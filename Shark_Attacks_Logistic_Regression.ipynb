```python
import pandas as pd
from pathlib import Path
```

```python
data = Path('shark_attacks_cleaned.csv')
df = pd.read_csv(data)
df.head()
```

```python
df = df.dropna()
df
```

```python
df['Fatal'] = df['Fatal'].str.replace('UNKNOWN','0')
df['Fatal'] = df['Fatal'].str.replace('N','0')
df['Fatal'] = df['Fatal'].str.replace('Y','1')
df
```

```python
X = pd.get_dummies(df, columns=['Date','Type','Country','Area','Location','Activity','Gender','Age'])
X = X.drop(columns=['Year','Fatal','Time','href formula','ISO_code'], axis=1)
y = df['Fatal']
```

```python
X.describe()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, predictions))
```
